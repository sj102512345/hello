```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('real_estate_price_size_year_view.csv')
df.head()
```

```python
df.dropna()
```

```python
df.describe()
```

### Rug Plot 

그래프의 X 축에 데이터를 1차원으로 표시

```python
sns.rugplot(df['size'])
plt.show()
```

### Kernel Density Plot

커널이라는 함수를 겹치는 방법으로 히스토그램보다 부드러운 형태의 분포 곡선을 보여주는 방법

```python
sns.kdeplot(df['price'])
plt.show()
```

### Dist Plot

러그와 커널 밀도 표시 기능이 있어서 Matplotlib의 hist 명령보다 많이 사용됨.

```python
sns.distplot(df['price'],kde=False)
plt.show()
```

```python
sns.distplot(df['price'], hist=False)
plt.show()
```

```python
sns.distplot(df['price'])
plt.show()
```

### Joint Plot

스캐터 플롯을 그리기 위해서는 jointplot 명령을 사용

```python
sns.jointplot(x='price',y='size',data=df)
plt.show()
```

```python
sns.jointplot(x='price',y='size',data=df,kind='hex')
plt.show()
```

kind='kde'이면 커널 밀도 히스토그램을 그림.

```python
sns.set_style('ticks')
sns.pairplot(df,hue='view',diag_kind='kde',kind='scatter',palette='husl')
plt.show()
```

```python
sns.stripplot(x='price',y='size',data=df)
plt.show()
```

```python
sns.stripplot(x='year',y='price',data=df,jitter=True)
plt.show()
```

```python
sns.swarmplot(x='view',y='price',data=df)
plt.show()
```

```python
sns.swarmplot(x="price", y="size", data=df)
plt.show()
```

```python
sns.violinplot(x="view",y='price',data=df)
plt.show()
```

```python
sns.violinplot(x="year",y="price",hue="view",data=df)
plt.show()
```

```python
sns.barplot(x="year",y="price",hue="view",data=df)
plt.show()
```

### Count Plot

각 카테고리 값별로 데이터가 얼마나 있는지 표시할 수 있음

```python
sns.countplot(x="year",data=df,palette="Blues");
plt.show()
```

```python
sns.pointplot(x="view",y="size",hue="year",data=df)
plt.show()
```

```python
sns.boxplot(data=df,orient="h")
plt.show()
```

### Cat Plot

색상(hue)과 행(row) 등을 동시에 사용하여 3 개 이상의 카테고리 값에 의한 분포 변화를 보여줌.

```python
sns.catplot(x="year",y="size",hue="view",data=df);
plt.show()
```

```python
sns.catplot(x="year",y="size",hue="view",kind='violin', data=df);
plt.show()
```

```python
sns.regplot(x="size",y="price",data=df)
sns.lmplot(x="size",y="price",data=df)
```

```python
g = sns.FacetGrid(df,col='view')
g.map(plt.hist,"size")
plt.show()
```

```python
g= sns.FacetGrid(df, col="price",hue="view")
g.map(plt.scatter, "price",'view')
plt.show()
```

```python
sns.set_palette("Set1")
g = sns.PairGrid(df, hue='size')
g.map(plt.scatter);
plt.show()
```

```python
sns.set_palette("Set2")
g= sns.PairGrid(df)
g.map_diag(plt.hist)
g.map_offdiag(plt.scatter);
plt.show()
```

```python
g = sns.PairGrid(df)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot,cmap="Blues_d")
g.map_diag(sns.kdeplot,lw=3, legend=False);
plt.show()
```

### Pair Plot

만약 3차원 이상의 데이터라면 seaborn 패키지의 pairplot 명령을 사용함. 
pairplot은 데이터프레임을 인수로 받아 그리드(grid) 형태로 각 데이터 열의 조합에 대해 스캐터 플롯을 그림. 

```python
sns.pairplot(df)
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.regplot(x = 'size', y = 'price', data = df)
plt.show()
```

```python
plt.figure(figsize=(20, 12))
plt.subplot(2,3,2)
sns.boxplot(x = 'year', y = 'price', data = df)
plt.subplot(2,3,3)
sns.boxplot(x = 'view', y = 'price', data = df)
plt.show()
```
